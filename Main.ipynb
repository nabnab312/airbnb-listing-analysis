```python
data.isnull().sum()

# dropping all missing value rows
data.dropna(inplace=True)

# data.fillna()
data.isnull().sum()
```

```python
data.shape
```

```python
#dealing with duplicates rows
data.duplicated().sum()


#deleting all duplicated rows
data[data.duplicated()]

data.drop_duplicates(inplace=True)
data.duplicated().sum()
```
